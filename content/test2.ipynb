```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pyodide
```

```python
url = "https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/drinks.csv"
data_url = pyodide.http.open_url(url)
```

```python
df = pd.read_csv(data_url)
df.head()
```

```python
df.groupby('continent').mean().plot(kind='bar');
```
